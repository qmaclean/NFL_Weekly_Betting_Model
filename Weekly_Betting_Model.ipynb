```python
current_df = current_df[["Week","Day","Date","Time","Winner/tie","H/A","Loser/tie","Pts Scored","Pts Allowed","Yds Gained","TOs","Yds Allowed","Opp TOs","Year"]]
```

```python
#final_df = historical_df.append(current_df)
final_df = current_df
```

```python
final_df['Score_Difference'] = final_df['Pts Scored'] - final_df['Pts Allowed']
```

```python
final_df['Total_Points'] = final_df['Pts Scored'] + final_df['Pts Allowed']
```

```python
final_df['home_win'] = np.where(final_df['H/A'] != '@',1,0)
```

```python
final_df['home_loss'] = np.where(final_df['H/A'] == '@',1,0)
```

```python
final_df.tail()
```

```python
final_df['visitor'] = np.where((final_df['home_loss'] == 0),final_df['Loser/tie'],
                               np.where((final_df['home_loss'] == 1),final_df['Winner/tie'],0))
```

```python
final_df['home'] = np.where((final_df['home_loss'] == 1),final_df['Loser/tie'],
                               np.where((final_df['home_win'] == 1),final_df['Winner/tie'],0))
```

```python
final_df.head()
```

```python
df_visitor = pd.get_dummies(final_df['visitor'],dtype=np.int64)
```

```python
df_home = pd.get_dummies(final_df['home'],dtype=np.int64)
```

```python
df_model = df_home.sub(df_visitor)
```

```python
df_model['Score_Difference'] = final_df['Score_Difference']
```

```python
df_model
```

```python
df_model_a = df_home.sub(df_visitor)
```

```python
df_model_a['Total_Points'] = final_df['Total_Points']
```

```python
df_model_a
```

```python
df_train = df_model # not required but I like to rename my dataframe with the name train.

lr = Ridge(alpha=0.001) 
X = df_train.drop(['Score_Difference'], axis=1)
y = df_train['Score_Difference']

lr.fit(X, y)
```

```python
#test versus actuals & time frames;
```

```python
df_ratings = pd.DataFrame(data={'team': X.columns,'rating': lr.coef_})
```

```python
df_ratings
```

```python
df_ratings.to_csv(f"df_ratings.csv", index=False)
```
